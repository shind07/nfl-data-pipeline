### Useful Columns
Since nflfastR features all the columns in nflscrapR, the nflscrapR documentation can be used to view most of the columns in nflfastR.

Below is an incomplete list of frquently used columns:

- posteam - the offensive team (possesion team)
- defteam - the defensive team
- game_id - a unique id given to each NFL game
- epa - expected points added
- wp - current win probability of the posteam
- def_wp - current win probability of the defteam
- yardline_100 - number of yards from the opponent's endzone
- passer - the player that passed the ball (on QB scramble plays, the QB is marked as a passer, not rusher)
- rusher - the player that ran the ball
- receiver - the player that was targeted on a pass
- passer_id, rusher_id, receiver_id - the player ID for the passer, rusher, or receiver on a play (useful for joining roster data)
- cpoe - completion percentage over expected on a given pass
- down - down of the play
- play_type - either run, pass, or no_play
- series_success - marked as a 1 if the series becomes successful (first down or a touchdown)

```python
df_pass_pbp_filtered = df_pass_pbp.loc[~((df_pass_pbp["incomplete_pass"] == 1) & (df_pass_pbp["complete_pass"] == 1))]
df_pass_pbp_filtered.shape
```

```python
cols = [
    "passer", "yards_gained", "pass_attempt", "pass_touchdown",
    "air_yards"
]

df_pass_pbp_complete = df_pass_pbp_filtered.loc[(df_pass_pbp_filtered["complete_pass"] == 1)]

df_pass_stats_complete = df_pass_pbp_complete[cols].groupby(["passer"], as_index=False).sum()
df_pass_stats_complete.sort_values("yards_gained", ascending=False).head(20)
```

```python
rename = lambda x: "complete_" + x if x not in ("passer", "interception", "yards_gained", "pass_touchdown") else x
df_pass_stats_complete.rename(rename, inplace=True, axis=1)
df_pass_stats_complete.head()
```

```python
cols = [
    "passer", "pass_attempt", "interception","air_yards"
]

df_pass_pbp_incomplete = df_pass_pbp_filtered.loc[
    (df_pass_pbp_filtered["incomplete_pass"] == 1) | (df_pass_pbp_filtered["interception"] == 1)
]

df_pass_stats_incomplete = df_pass_pbp_incomplete[cols].groupby(["passer"], as_index=False).sum()
df_pass_stats_incomplete.sort_values("air_yards", ascending=False).head(20)
```

```python
rename = lambda x: "incomplete_" + x if x not in ("passer", "interception") else x
df_pass_stats_incomplete.rename(rename, inplace=True, axis=1)
df_pass_stats_incomplete.head()
```

```python
df_pass_stats = df_pass_stats_complete.merge(df_pass_stats_incomplete, on=['passer'])

def format_attempts_completions(row):
    attempts = row['complete_pass_attempt'] + row['incomplete_pass_attempt']
    return f"{int(row['complete_pass_attempt'])} / {int(attempts)}"

def format_percentage(column):
    formatted_column = int(round(column))
    return str(formatted_column) + '%'

def get_fantasy_points(row):
    return \
        (row['pass_touchdown'] * 4) \
        - (row['interception'] * 2) \
        + (row['yards_gained']) / 25
    
df_pass_stats['Comp/Att'] = df_pass_stats.apply(format_attempts_completions, axis=1)
df_pass_stats['Comp%'] = df_pass_stats['complete_pass_attempt'] * 100 /(df_pass_stats['complete_pass_attempt'] + df_pass_stats['incomplete_pass_attempt'])
df_pass_stats['Comp%'] = df_pass_stats['Comp%'].apply(format_percentage)

df_pass_stats['Att. AY'] = df_pass_stats['complete_air_yards'] + df_pass_stats['incomplete_air_yards']
df_pass_stats['AY Comp%'] = df_pass_stats['complete_air_yards'] * 100 / (df_pass_stats['complete_air_yards'] + df_pass_stats['Att. AY'])
df_pass_stats['AY Comp%'] = df_pass_stats['AY Comp%'].apply(format_percentage)

df_pass_stats['AY%'] = 100 * df_pass_stats['complete_air_yards'] / df_pass_stats['yards_gained']
df_pass_stats['AY%'] = df_pass_stats['AY%'].apply(format_percentage)

df_pass_stats['aDOT'] = df_pass_stats['Att. AY'] / \
    (df_pass_stats['complete_pass_attempt'] + df_pass_stats['incomplete_pass_attempt'])
df_pass_stats['aDOT'] = df_pass_stats['aDOT'].round(1)

df_pass_stats['Y/A'] = df_pass_stats['yards_gained'] / \
    (df_pass_stats['complete_pass_attempt'] + df_pass_stats['incomplete_pass_attempt'])
df_pass_stats['Y/A'] = df_pass_stats['Y/A'].round(1)

df_pass_stats['TD/INT'] = df_pass_stats.apply(lambda x: str(int(x['pass_touchdown'])) + ' / ' + str(int(x['interception'])), axis=1)

df_pass_stats['Fantasy Pts'] = df_pass_stats.apply(get_fantasy_points, axis=1)

df_pass_stats['passer'] = df_pass_stats['passer'].apply(lambda x: x.replace('.', ', '))

df_pass_stats
```

```python
for col, dtype in dict(df_pass_stats.dtypes).items():
    if col in ('aDOT', 'Y/A', 'Fantasy Pts'): continue
    if dtype == np.float64: 
        df_pass_stats[col] = df_pass_stats[col].astype(int)
```
